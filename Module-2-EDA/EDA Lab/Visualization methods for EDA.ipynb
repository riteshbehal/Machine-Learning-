```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('HR-Employee-Attrition.csv')
df.head()
```

```python
df.info()
```

```python
# Univariate
# Histogram
sns.histplot(data=df,x='Age')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(data=df,x='Attrition',color='blue')
```

```python
df['Attrition'].value_counts(normalize=True)
```

```python
sns.countplot(data=df,x='Department',hue='Attrition')
```

```python
# 1 is numerical & another is categorical
sns.barplot(data=df,x='Attrition',y='MonthlyIncome')
plt.show()
```

```python
# 2 continuous
sns.scatterplot(data=df,x='MonthlyIncome',y='YearsAtCompany',hue='Attrition')
plt.show()
```

```python
# swarm plot
# scatter plot for categorical and numerical data
plt.figure(figsize=(12,8))
sns.swarmplot(data=df, y='JobRole',x='MonthlyIncome',hue='Attrition')
plt.show()
```

```python
sns.boxplot(x='Department',y='MonthlyIncome',data=df)
plt.show()
```

```python
sns.jointplot(data=df, x='MonthlyIncome',y='Age',kind='scatter')
```
